```python
# cleaning 
# i created a new df with just relevant features pertaining to collisions seen and weather affects
# used the loc function, : got all rows, and '' to get relevant features
df1 = collisions.loc[:,['count','uwind', 'vwind', 'ceiling_height', 'visibility_distance', 'visibility', 'temperature', 'amt', 'scaled_yday', 'scaled_uwind', 'scaled_vwind', 'scaled_ceiling_height' ]]
```

```python
df1
```

```python
# used boolean series to get all the observations where the count of collisions was above 0
# this gave us only observations where collisions happened
# i removed all observations where collisions were not seen (count=0) and named it filtered_df
filtered_df = df1[df1['count'] > 0]
```

```python
# 3.3
import seaborn as sns #(need to do this to make visualization)
#use a histogram (sns.displot) to display temperature against count of collisions with filtered_df
# set data to filtered_df and x will be temp in string and y will be count in string
sns.displot(data=filtered_df, x = 'temperature', y= 'count')
# this plot is appropriate because it arranges the temps into bins and will show frequency of how many collisions fall into those bins
# this will ultimately tell you in what temps the most bird collisions were seen in
```

```python
# in the plot above it plots the same mean as shown in the describe function, as well as the max and min shown by
# filtered_df['temperature'].describe(), from the plot it is shown that in higher temps there are more collisions
# this evidence contributes to the papers hypothesis because it proves that there are more bird collisions in worse weather conditions
```

```python
# use sns.catplot function, set data to filter_df and x to amt in string and y to count in string
# this plot is useful beacuse it shows each observation as a point which describes a single collision in low or high amt
# amt is average migration traffic, which means if there were many birds migrating as oppose to few
# the plot shows the same values as filtered_df['amt'].describe() function although the x axis does not display values 
# with this visualization you can see that in higher migrational traffic there are more bird collisions seen
# this supports the hypothesis that more AMT contributes to more bird collisions into buildings
sns.catplot(data=filtered_df, x = 'amt', y= 'count')
# used a scatterplot to display average migration traffic against count of collisions happening
```
